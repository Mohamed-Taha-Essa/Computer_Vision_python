```python
import numpy as np
import cv2

%matplotlib inline
# Read in the image
cap= cv2.VideoCapture("Woman.mp4")
## TODO: Define our color selection boundaries in RGB values
lower_green = np.array([0,120,0]) 
upper_green = np.array([100,255,100])
while(cap.isOpened()):
    ret,image=cap.read()
    if ret:
        image =cv2.resize(image,(720,575))
        # Define the masked area
        #back ground white and object is black
        mask = cv2.inRange(image, lower_green, upper_green)
        
        # Mask the image to let the car show through
        #back ground in image will be black
        masked_image = np.copy(image)
        masked_image[mask != 0] = [0, 0, 0]
        # Load in a background image, and convert it to RGB 
        background_image = cv2.imread('images/sky.jpg')
        ## TODO: Crop it or resize the background to be the right size (450x660)
        crop_background=background_image[0:575,0:720]

        ## TODO: Mask the cropped background so that the car area is blocked
        crop_background[mask == 0] = [0,0,0]
        ## TODO: Add the two images together to create a complete image!
        complet_image=masked_image +crop_background

        
        cv2.imshow("image",image)
        cv2.imshow("mask",mask)
        cv2.imshow("masked_image",masked_image)
        cv2.imshow("crop_background",crop_background)
        cv2.imshow("complet_image",complet_image)


        key =cv2.waitKey(10)
        if (key==27):
            break
        
    else:
        break
        
cap.release() 
cv2.destroyAllWindows()
```
